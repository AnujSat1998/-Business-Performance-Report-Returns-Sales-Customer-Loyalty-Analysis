```python
# Function to standardize categorical values
def clean_text(value):
    if isinstance(value, str):  # Check if value is a string
        return value.strip()  # Remove leading & trailing spaces
    return value  # Return original value if not a string

# Clean spaces from all categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].apply(clean_text)

# Standardize specific categorical columns
mapping_dict = {
    'income_bracket': {'L': 'Low', 'M': 'Medium', 'H': 'High'},
    'app_usage': {'H': 'High', 'M': 'Medium', 'L': 'Low'},
    'social_media_engagement': {'H': 'High', 'M': 'Medium', 'L': 'Low'}
}

for col, mapping in mapping_dict.items():
    df[col] = df[col].replace(mapping)

# Save cleaned dataset
updated_file_path = "final_cleaned_retail_data.csv"
df.to_csv(updated_file_path, index=False)

# Print confirmation message
print(f"Data cleaned and saved as: {updated_file_path}")
```

```python
import pandas as pd
from datetime import datetime, timedelta

# Load CSV file
file_path = 'final_standardized_retail_data.csv'
df = pd.read_csv(file_path)

# List of date columns to convert
date_cols = ['transaction_date', 'last_purchase_date', 'promotion_start_date',
             'promotion_end_date', 'product_expiry_date', 'product_manufacture_date']

# Function to correct invalid day values (e.g., 4/31 â†’ 4/30)
def fix_invalid_date(date_obj):
    if pd.isna(date_obj):
        return pd.NaT  # If it's NaT, return as is

    while True:
        try:
            # Convert to valid date (if invalid, it will raise ValueError)
            valid_date = datetime.strptime(str(date_obj), "%Y-%m-%d").date()
            return valid_date  # If valid, return
        except ValueError:
            # Reduce day by 1 if invalid (e.g., 31st in April â†’ 30th)
            date_obj -= timedelta(days=1)

# Function to parse & validate dates
def safe_parse_date(date_str):
    try:
        # Convert to datetime, coerce errors (invalid dates â†’ NaT)
        date_obj = pd.to_datetime(date_str, errors='coerce', dayfirst=False).date()

        # Drop rows where the month is greater than 12
        if date_obj and date_obj.month > 12:
            return None  # Mark for deletion

        # Apply correction for invalid day values
        return fix_invalid_date(date_obj)

    except Exception:
        return pd.NaT  # Return NaT if all else fails

# Apply function to all date columns
for col in date_cols:
    df[col] = df[col].apply(safe_parse_date)

# Drop rows where any date column is `None` (invalid months > 12)
df.dropna(subset=date_cols, inplace=True)

# Fill missing values (`NaT`) using the previous valid date
df.fillna(method='ffill', inplace=True)

# Save the updated DataFrame to an Excel file
output_file = 'updated_file.xlsx'
df.to_excel(output_file, index=False, engine='openpyxl')

print(f"âœ… Date columns have been standardized, invalid months removed, and saved to {output_file}")
```

```python
import pandas as pd

# Load dataset
file_path = 'updated_file.xlsx'  # Replace with your file's name and extension
df = pd.read_excel(file_path)

# Convert transaction_date to datetime format
df['transaction_date'] = pd.to_datetime(df['transaction_date'], errors='coerce')

# Check the minimum and maximum dates
min_date = df['transaction_date'].min()
max_date = df['transaction_date'].max()

print(f"ðŸ“… Date Range in Dataset: {min_date} to {max_date}")
```

```python
# Group by Quarter and Sum Total Sales
sales_by_quarter = df.groupby(df['transaction_date'].dt.to_period("Q"))['total_sales'].sum()

# Print Sales Trends
print("ðŸ“‰ Total Sales Per Quarter:\n", sales_by_quarter)

# Count Transactions Per Quarter
transaction_counts = df.groupby(df['transaction_date'].dt.to_period("Q"))['transaction_id'].count()

# Print Transaction Trends
print("ðŸ›’ Transactions Per Quarter:\n", transaction_counts)

# Calculate Average Order Value (AOV) Per Quarter
df['order_value'] = df['quantity'] * df['unit_price']
aov_by_quarter = df.groupby(df['transaction_date'].dt.to_period("Q"))['order_value'].mean()

# Print AOV Trends
print("ðŸ’° Average Order Value Per Quarter:\n", aov_by_quarter)

# Extract Year and Quarter
df['year'] = df['transaction_date'].dt.year
df['quarter'] = df['transaction_date'].dt.quarter

# Compare Q3 of Each Year
q3_sales = df[(df['quarter'] == 3)].groupby('year')['total_sales'].sum()
print("ðŸ“Š Q3 Sales Comparison (2020 vs. 2021):\n", q3_sales)
```
